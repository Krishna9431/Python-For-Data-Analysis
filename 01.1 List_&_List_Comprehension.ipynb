### List Method

```python
fruits = ["apple", "banana", "cherry", "mango"]

# append a new item
fruits.append("orange")
print("Append 'orange', the list is now:",fruits)

```

```python
# insert a fruits in index value 1
fruits.insert(1,"banana") # here its not replacing the value, its inserting the value in index 1
print(fruits)
```

```python
# Remove an item in the list
fruits.remove("banana") # Removing first occurance of the item 
print(fruits)

```

```python
# Remove and return the last element
popped_fruit = fruits.pop()
print(popped_fruit)
print(fruits)
```

```python
# Hoe to get index of the list item
print(fruits.index("banana"))
print(fruits.index("mango"))
```

```python
# sorting list

unsorted_list = ["Hello", "World", "Python", "Programming"]
unsorted_list.sort()
unsorted_list
```

```python
# Reverse

fruits.reverse()
print(fruits)

```

```python
# Remove all the item from the list
fruits.clear()
fruits
```

### Iterating List

```python
for i in x:
    print(i)
```

```python
# using enumerate
for idx, i in enumerate(x):
    print(f"Index = {idx} -> element = {i}")
```

### List Comprehension

- Basic Syntex -> `[expression for item in iterable]`
- Cinditional logic - `[expression for item in iterable if condition]`
- Nested list comprehension - `[expression for item1 in iterable1 for item2 in iterable2]`



```python
# Normal code
lst = []
for x in range(11):
    lst.append(x**2)

print(lst)
```

```python
# using comprehension
[x**2 for x in range(11)]
```

```python
# Q. Squre of number
sqaure = [num**2 for num in range(11)]
print(sqaure)
```

```python
# Square of even numbers
even_sqaure = [i**2 for i in range(11) if i%2==0]
print(even_sqaure)
```

```python
# Nested list comprehension

lst1= [1,2,3,4]
lst2= ['a', 'b', 'c', 'd']

pair = [[i,j] for i in lst1 for j in lst2]
print(pair)
```

```python
# With function calls

words = ["Apple", "Krishna", "samsung", "constant"]
lengths = [len(word) for word in words]
print(lengths)
```
