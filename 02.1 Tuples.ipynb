# Tuples
- Tuples are ordered collection of items that are immutable. They are simillar to list, but their immutability makes them different.
- eg - Empty_tuple = ()

```python
# creating tuples
x =()
print(x)
print(type(x))


```

```python
lst = list()
tup = tuple()

print(type(lst))
print(type(tup))
```

```python
# Type conversion

a = tuple([1,3,5,7]) # Convert list to tuple
print(type(a))

b = list((1,3,4,6,8)) # convert tuple into list
print(type(b))
```

```python
mixed_tuples = (1, "and", True, 3.54)
print(mixed_tuples)
for i in mixed_tuples:
    print(type(i))
```

```python
# Access tuple element, same as list
num = (2,7,95,3,30)
print(num[2])
print(num[-1])
print(num[-2])
print(num[1:4])
print(num[:4])
```

```python
# Operations

print(num)
print(mixed_tuples)

concatenation_tuple = num + mixed_tuples
print(concatenation_tuple)
print(type(concatenation_tuple))
```

```python
num * 3
```

### Immutable nature of tuple
- means their element cannot be changed once assigned.

```python
# List
l1 = [1,23,4,6,8]
print(l1)

l1[2] = "Krishna" # It replace the 2nd index value with 'krishna'
print(l1)
```

```python
# Using in tuple
t1 = (1,3,5,7,5)
print(t1)
t1[2] = "Krishna"  # TypeError: 'tuple' object does not support item assignment
print(t1)

# If you want to do that of operation then first convert tuple -> list and then List -> tuple
```

```python
# Methods

# counts()
# syntex = tuple_name.count(element)
print(num)
print(num.count(1)) # its search 1 that is present or not? if present the count how many times its present
print(num.count(2)) # its search 2 that is present or not? if present the count how many times its present
```

```python
# index()
# syntex -> tuple_name.index(element)   #Its find the index value that the element is presrent

print(num)
print(num.index(2))
print(num.index(30))
print(num.index(300)) # ValueError: tuple.index(x): x not in tuple
```

#### Packing and unpacking tuple

```python
packed_tuple= 1,"Krishna", 3.14
print(packed_tuple)
```

```python
# unpacking
a,b,c = packed_tuple
print(a)
print(b)
print(c)
```

```python
# unpacking with *

num2 = (1,5,3,6,8,9,0,6,4,3,12)
first, *middle, last = num2
print(first) # its take the first value
print(middle) # its take the middle value except first and last
print(last) # its take the last value
```

### Nested tuple

```python
# Example of tuples in the list

lst1 = [[1,4,3], ["A", 45, "t"], ("Hello", 34, 67)]
print(lst1[0])
print(lst1[0][2])
print(lst1[0][:2])

print(lst1[2])
print(lst1[2][2])
print(lst1[2][:2])

```

```python
# Tuples in tuples
tup1 = ((1,4,3), ("A", 45, "t"), ("Hello", 34, 67))
print(tup1[0])
print(tup1[0][2])
print(tup1[0][:2])

print(tup1[2])
print(tup1[2][2])
print(tup1[2][:2])
```

```python
for sub_tuple in tup1:
    for item in sub_tuple:
        print(item, end=" ")
    print()
```
